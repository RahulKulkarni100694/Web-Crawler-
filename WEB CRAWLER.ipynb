```python
######### main program ###########
A1 = []
source = []
author = []
date_time = []
html = driver.page_source
soup = BeautifulSoup(html, "lxml")
g_data2 = soup.find_all("div", {"class": "leadFields"})
def extract():

    global count
    count = countvalue()
    for i in range(77,count):  
        A2=driver.find_elements_by_class_name('enHeadline')
        driver.get(str(A2[i].get_attribute("href")))
        time.sleep(15)
        #continue
        
        le = len(driver.find_elements_by_class_name('wsj-snippet-login'))
        le1 = len(driver.find_elements_by_class_name('article-body'))
        le2 = len(driver.find_elements_by_class_name('mceTemp'))

        
        if le > 0:
            driver.execute_script("window.history.go(-1)")
            time.sleep(10)
            continue
          
        if le1 > 0:
            driver.execute_script("window.history.go(-1)")
            time.sleep(10)
            continue
            
        if le2 > 0:
            driver.execute_script("window.history.go(-1)")
            time.sleep(10)
            continue
           
            
        A2=driver.find_element_by_class_name('article').text
        A1.append(A2)
        #A2=driver.find_element_by_class_name('article-body').text
        #A1.append(A2)
        
                        #print(A1)
        print(i)
        print('-------------------------------------------')
                        # print(driver.page_source)
        driver.execute_script("window.history.go(-1)")
        time.sleep(15)
        a = g_data2[i].text
        b =a.split(',')[0]
        source.append(b)
        c=a.split('words,')[1]
        author.append(c)
        d = a.split(',')[ 1:  ]
        str1 = ','.join(d)
        z = re.findall(r"(\d+) words",str1)
        w = ''.join(z)
        x =', '+ w+' words,'
        p = str1.split(x)[0]
        date_time.append(p)
        time.sleep(8) 
        if i == count -1:
            next = driver.find_element_by_class_name('nextItem')
            if driver:
                next.click()
                time.sleep(20)
                count = countvalue()
                if count >0:
                    extract()
```
